```python
# winsorization set the values of the lowest end to the value of 
# the 5th percentile.Let's winsorize the life_expectancy variable:

from scipy.stats.mstats import winsorize

# We apply one-way winsorization to the highest end
# of the value distribution. As we have a lot of extreme
# values, we set the values of the outliers to the 
# value of the 90th percentile. 
winsorized_views = winsorize(life_df1_['life_expectancy'], (0.10, 0))

winsorized_views
```

```python
plt.boxplot(winsorized_views)
plt.title("Box plot of life_expectancy (whis=1.5)")
plt.show()
```
